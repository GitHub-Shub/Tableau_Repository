```python
gender_counts = df.groupby('Gender').size()

# Create the bar chart
colors = ['blue', 'orange','black']
plt.figure(figsize=(10,5))
gender_counts.plot(kind='bar', title='Gender Ratio',color=colors)

# Annotate each bar with its corresponding count
for i, (gender, count) in enumerate(gender_counts.items()):
    plt.annotate(f'{count}', xy=(i, count), ha='center', va='bottom')

# Set labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Ratio')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```
